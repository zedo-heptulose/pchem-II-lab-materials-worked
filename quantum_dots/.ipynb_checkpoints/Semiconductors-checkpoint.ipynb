```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
```

```python
def plot_band_structure(bandgap=1.5, effective_mass_ratio=0.5, is_direct=True):
    """Plot simplified band structure for direct/indirect bandgap semiconductors"""
    k = np.linspace(-2, 2, 1000)
    
    # Valence band (negative effective mass)
    E_v = -k**2/(2*effective_mass_ratio)
    
    # Conduction band
    if is_direct:
        E_c = bandgap + k**2/2  # Direct bandgap
    else:
        offset = 0.7  # k-space offset for indirect bandgap
        E_c = bandgap + (k-offset)**2/2  # Indirect bandgap
    
    plt.figure(figsize=(8, 6))
    plt.plot(k, E_v, 'b-', linewidth=2, label='Valence Band')
    plt.plot(k, E_c, 'r-', linewidth=2, label='Conduction Band')
    
    # Highlight the bandgap
    if is_direct:
        plt.annotate('', xy=(0, 0), xytext=(0, bandgap),
                    arrowprops=dict(arrowstyle='<->', color='green', lw=2))
        plt.text(0.1, bandgap/2, f'Eg = {bandgap:.1f} eV', color='green')
    else:
        plt.annotate('', xy=(offset, 0), xytext=(offset, bandgap),
                    arrowprops=dict(arrowstyle='<->', color='green', lw=2))
        plt.text(offset+0.1, bandgap/2, f'Eg = {bandgap:.1f} eV', color='green')
        
        # Show photon transition for indirect bandgap
        plt.annotate('', xy=(0, 0), xytext=(offset, bandgap),
                    arrowprops=dict(arrowstyle='->', linestyle='--', color='purple', lw=1.5))
        plt.text(offset/2-0.3, bandgap/2, 'phonon\nassisted', color='purple')
    
    plt.axhline(y=0, color='gray', linestyle='--', alpha=0.7)
    plt.axhline(y=bandgap, color='gray', linestyle='--', alpha=0.7)
    
    plt.xlabel('Crystal Momentum (k)')
    plt.ylabel('Energy (eV)')
    plt.title('Band Structure: ' + ('Direct' if is_direct else 'Indirect') + ' Bandgap Semiconductor')
    plt.legend(loc='best')
    plt.grid(True, alpha=0.3)
    plt.ylim(-1, bandgap+1)
    plt.show()

def quantum_confinement_effect(bulk_bandgap=1.5, effective_mass=0.1, dielectric_constant=10):
    """Visualize quantum confinement effect on bandgap"""
    # Calculate exciton Bohr radius
    m0 = 9.11e-31  # electron mass in kg
    hbar = 1.054e-34  # reduced Planck constant
    e = 1.602e-19  # elementary charge
    epsilon0 = 8.85e-12  # vacuum permittivity
    
    # Reduced mass (assuming electron and hole have same effective mass)
    mr = effective_mass * m0 / 2
    
    # Exciton Bohr radius in nm
    bohr_radius = 4*np.pi*dielectric_constant*epsilon0*hbar**2/(e**2*mr) * 1e9
    
    # Quantum dot sizes
    sizes = np.linspace(1, 20, 100)  # nm
    
    # Calculate bandgap with quantum confinement
    confinement_energy = (hbar**2 * np.pi**2)/(2 * mr) * (1/(sizes*1e-9)**2) * 6.242e18  # convert J to eV
    bandgap = bulk_bandgap + confinement_energy
    
    plt.figure(figsize=(10, 6))
    plt.plot(sizes, bandgap, 'b-', linewidth=2)
    plt.axhline(y=bulk_bandgap, color='gray', linestyle='--', 
                label=f'Bulk Bandgap = {bulk_bandgap} eV')
    plt.axvline(x=bohr_radius, color='red', linestyle='--', 
                label=f'Exciton Bohr Radius = {bohr_radius:.1f} nm')
    
    # Annotate regions
    plt.text(bohr_radius/2, max(bandgap)*0.9, 'Strong\nConfinement', 
             ha='center', color='darkred')
    plt.text(bohr_radius*1.5, bulk_bandgap*1.1, 'Weak\nConfinement', 
             ha='center', color='darkblue')
    
    plt.xlabel('Quantum Dot Diameter (nm)')
    plt.ylabel('Bandgap Energy (eV)')
    plt.title('Quantum Confinement Effect on Semiconductor Bandgap')
    plt.ylim(bulk_bandgap*0.9, max(bandgap)*1.1)
    plt.xlim(0, 20)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()
    
    return bohr_radius, bandgap

def interactive_material_explorer():
    """Interactive plot to explore semiconductor properties"""
    # Template for interactive visualization
    # Would implement sliders for bandgap, effective mass, etc.
    pass
```

```python
def plot_simple_band_structure(bandgap=1.5, band_width=1.0, band_states=30, 
                              fraction_filled=1.0, material_type="semiconductor"):
    """
    Plot simplified band structure showing valence and conduction bands
    
    Parameters:
    bandgap - Energy difference between bands (eV)
    band_width - Width of each band (eV)
    band_states - Number of states to show in each band
    fraction_filled - Fraction of valence band that is filled (0-1)
    material_type - "insulator", "semiconductor", or "conductor"
    """
    plt.figure(figsize=(8, 6))
    
    # Set up parameters based on material type
    if material_type == "insulator":
        bandgap = 5.0
        fraction_filled = 1.0
    elif material_type == "semiconductor":
        # Use provided bandgap
        fraction_filled = 1.0
    elif material_type == "conductor":
        bandgap = 0.0
        fraction_filled = 0.7
    
    # Create energies for valence band (evenly spaced)
    valence_energies = np.linspace(-band_width, 0, band_states)
    
    # Create energies for conduction band (evenly spaced)
    conduction_energies = np.linspace(bandgap, bandgap + band_width, band_states)
    
    # Calculate Fermi level
    if material_type == "conductor":
        fermi_level = valence_energies[int(fraction_filled * band_states)]
    else:
        fermi_level = valence_energies[-1] + bandgap/2
    
    # Calculate which states are occupied
    valence_occupied = valence_energies <= fermi_level
    conduction_occupied = conduction_energies <= fermi_level
    
    # Plot bands
    # Valence band
    plt.scatter(np.zeros_like(valence_energies[valence_occupied]), 
                valence_energies[valence_occupied], 
                color='gold', s=70, edgecolor='black', zorder=3, label="Occupied")
    
    plt.scatter(np.zeros_like(valence_energies[~valence_occupied]), 
                valence_energies[~valence_occupied], 
                color='black', s=70, edgecolor='black', zorder=3)
    
    # Conduction band
    plt.scatter(np.zeros_like(conduction_energies[conduction_occupied]), 
                conduction_energies[conduction_occupied], 
                color='gold', s=70, edgecolor='black', zorder=3)
    
    plt.scatter(np.zeros_like(conduction_energies[~conduction_occupied]), 
                conduction_energies[~conduction_occupied], 
                color='black', s=70, edgecolor='black', zorder=3, label="Unoccupied")
    
    # Add band labels
    plt.text(0.05, valence_energies.mean(), "Valence\nBand", ha='left', va='center')
    plt.text(0.05, conduction_energies.mean(), "Conduction\nBand", ha='left', va='center')
    
    # Label the bandgap
    if bandgap > 0:
        plt.annotate('', xy=(0, conduction_energies[0]), xytext=(0, valence_energies[-1]),
                    arrowprops=dict(arrowstyle='<->', color='red', lw=2))
        plt.text(0.05, valence_energies[-1] + bandgap/2, f"Bandgap\n{bandgap:.1f} eV", 
                 ha='left', va='center', color='red')
    
    # Label Fermi level
    plt.axhline(y=fermi_level, color='blue', linestyle='--', alpha=0.7)
    plt.text(-0.1, fermi_level, "Fermi\nLevel", ha='right', va='center', color='blue')
    
    # Set labels and title
    plt.title(f"{material_type.capitalize()} Band Structure")
    plt.ylabel("Energy (eV)")
    plt.xlim(-0.2, 0.2)
    
    # Hide x-axis
    plt.tick_params(axis='x', which='both', bottom=False, labelbottom=False)
    
    # Add legend only if needed
    if not (fraction_filled == 1.0 or fraction_filled == 0.0):
        plt.legend(loc='upper right')
    
    plt.grid(True, axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()

def plot_quantum_confinement_effect(bulk_bandgap=1.5, max_diameter=20):
    """Visualize quantum confinement effect on bandgap"""
    # Quantum dot sizes
    sizes = np.linspace(2, max_diameter, 100)  # nm
    
    # Calculate bandgap with quantum confinement (simplified model)
    # Uses E = Ebulk + C/d
    confinement_constant = 2.0  # eVnm
    bandgap = bulk_bandgap + confinement_constant/(sizes**2)
    
    # Create figure
    plt.figure(figsize=(10, 6))
    
    # Plot bandgap vs size
    plt.plot(sizes, bandgap, 'b-', linewidth=3)
    plt.axhline(y=bulk_bandgap, color='gray', linestyle='--', 
                label=f'Bulk Bandgap = {bulk_bandgap} eV')
    
    # Add labels
    plt.xlabel('Quantum Dot Diameter (nm)')
    plt.ylabel('Bandgap Energy (eV)')
    plt.title('Quantum Confinement Effect on Semiconductor Bandgap')
    plt.grid(True, alpha=0.3)
    plt.legend()
    
    # Create insets showing band structures at different sizes
    # Small quantum dot
    small_size_idx = 10  # Index for a small QD
    small_size = sizes[small_size_idx]
    small_bg = bandgap[small_size_idx]
    
    # Large quantum dot
    large_size_idx = 80  # Index for a larger QD
    large_size = sizes[large_size_idx]
    large_bg = bandgap[large_size_idx]
    
    # Add band structure insets for both sizes
    from mpl_toolkits.axes_grid1.inset_locator import inset_axes
    
    # Function to create a minimal band diagram
    def add_mini_band_diagram(ax, bg, size, pos):
        # Draw valence and conduction bands
        ax.plot([-0.5, 0.5], [0, 0], 'k-', linewidth=2)
        ax.plot([-0.5, 0.5], [bg, bg], 'k-', linewidth=2)
        
        # Fill valence band
        ax.fill_between([-0.5, 0.5], [-0.2, -0.2], [0, 0], color='gold')
        
        # Label bandgap
        ax.text(0, bg/2, f"{bg:.2f} eV", ha='center', va='center')
        
        # Set title and hide axes
        ax.set_title(f"{size:.1f} nm", fontsize=9)
        ax.axis('off')
    
    # Create and position insets
    small_inset = inset_axes(plt.gca(), width="15%", height="25%", 
                             loc='upper right', bbox_to_anchor=(0.25, 0.9, 0.3, 0.3))
    add_mini_band_diagram(small_inset, small_bg, small_size, 'upper right')
    
    large_inset = inset_axes(plt.gca(), width="15%", height="25%", 
                              loc='lower right', bbox_to_anchor=(0.8, 0.15, 0.3, 0.3))
    add_mini_band_diagram(large_inset, large_bg, large_size, 'lower right')
    
    # Add arrows connecting points to insets
    plt.annotate('', xy=(small_size, small_bg), xytext=(small_size+3, small_bg+0.5),
                arrowprops=dict(arrowstyle='->', color='black', lw=1.5))
    
    plt.annotate('', xy=(large_size, large_bg), xytext=(large_size-3, large_bg+0.5),
                arrowprops=dict(arrowstyle='->', color='black', lw=1.5))
    
    plt.show()
```

```python
# Show band structures for different materials
print("Insulator Band Structure:")
plot_simple_band_structure(material_type="insulator")

print("Semiconductor Band Structure:")
plot_simple_band_structure(bandgap=1.5, material_type="semiconductor")

print("Conductor Band Structure:")
plot_simple_band_structure(material_type="conductor")

# Show quantum confinement effect
plot_quantum_confinement_effect(bulk_bandgap=1.5, max_diameter=20)
```

```python
def plot_simple_band_structure(bandgap=1.5, band_width=1.0, 
                              material_type="semiconductor"):
    """
    Plot simplified band structure showing bands with boxes
    
    Parameters:
    bandgap - Energy difference between bands (eV)
    band_width - Width of each band (eV)
    material_type - "insulator", "semiconductor", or "conductor"
    """
    plt.figure(figsize=(8, 6))
    
    # Set up parameters based on material type
    if material_type == "insulator":
        bandgap = 5.0
        valence_fill = 1.0
    elif material_type == "semiconductor":
        # Use provided bandgap
        valence_fill = 1.0
    elif material_type == "conductor":
        bandgap = 0.5  # Small gap for clarity
        valence_fill = 0.5  # Half-filled valence band
    
    # Box dimensions
    box_width = 0.4
    box_x = [-box_width/2, box_width/2]
    
    # Energy ranges for bands
    core_band = [-band_width*3 - 1, -band_width*2 - 1]
    valence_band = [-band_width, 0]
    conduction_band = [bandgap, bandgap + band_width]
    
    # Calculate fill heights
    valence_fill_height = valence_band[0] + (valence_band[1] - valence_band[0]) * valence_fill
    
    # Draw core band (always fully filled)
    plt.fill_between(box_x, [core_band[0], core_band[0]], [core_band[1], core_band[1]], 
                     color='gold', edgecolor='black', linewidth=1.5, zorder=3)
    
    # Draw valence band - filled portion
    plt.fill_between(box_x, [valence_band[0], valence_band[0]], [valence_fill_height, valence_fill_height], 
                     color='gold', edgecolor=None, linewidth=0, zorder=3)
    
    # Draw valence band - unfilled portion
    if valence_fill < 1.0:
        plt.fill_between(box_x, [valence_fill_height, valence_fill_height], [valence_band[1], valence_band[1]], 
                        color='black', alpha=0.8, edgecolor=None, linewidth=0, zorder=3)
    
    # Draw valence band outline
    plt.plot([box_x[0], box_x[0]], [valence_band[0], valence_band[1]], 'k-', linewidth=1.5, zorder=4)
    plt.plot([box_x[1], box_x[1]], [valence_band[0], valence_band[1]], 'k-', linewidth=1.5, zorder=4)
    plt.plot([box_x[0], box_x[1]], [valence_band[0], valence_band[0]], 'k-', linewidth=1.5, zorder=4)
    plt.plot([box_x[0], box_x[1]], [valence_band[1], valence_band[1]], 'k-', linewidth=1.5, zorder=4)
    
    # Draw conduction band (always empty for semiconductors and insulators)
    if material_type == "conductor":
        # In conductors, the bands overlap, so conduction band is partially filled
        plt.fill_between(box_x, [conduction_band[0], conduction_band[0]], [conduction_band[1], conduction_band[1]], 
                        color='black', alpha=0.8, edgecolor='black', linewidth=1.5, zorder=3)
    else:
        plt.fill_between(box_x, [conduction_band[0], conduction_band[0]], [conduction_band[1], conduction_band[1]], 
                        color='black', alpha=0.8, edgecolor='black', linewidth=1.5, zorder=3)
    
    # Add band labels
    plt.text(box_width*0.7, core_band[0] + (core_band[1] - core_band[0])/2, 
             "Core\nBand", ha='left', va='center')
    plt.text(box_width*0.7, valence_band[0] + (valence_band[1] - valence_band[0])/2, 
             "Valence\nBand", ha='left', va='center')
    plt.text(box_width*0.7, conduction_band[0] + (conduction_band[1] - conduction_band[0])/2, 
             "Conduction\nBand", ha='left', va='center')
    
    # Label the bandgap
    if bandgap > 0 and material_type != "conductor":
        plt.annotate('', xy=(0, conduction_band[0]), xytext=(0, valence_band[1]),
                    arrowprops=dict(arrowstyle='<->', color='red', lw=2))
        plt.text(box_width*0.7, valence_band[1] + bandgap/2, f"Bandgap\n{bandgap:.1f} eV", 
                 ha='left', va='center', color='red')
    
    # Calculate Fermi level
    if material_type == "conductor":
        fermi_level = valence_fill_height
    else:
        fermi_level = valence_band[1] + bandgap/2
    
    # Label Fermi level
    plt.axhline(y=fermi_level, color='blue', linestyle='--', alpha=0.7)
    plt.text(-box_width, fermi_level, "Fermi\nLevel", ha='right', va='center', color='blue')
    
    # Set labels and title
    plt.title(f"{material_type.capitalize()} Band Structure")
    plt.ylabel("Energy (eV)")
    plt.xlim(-box_width*2, box_width*2)
    
    # Adjust y limits to show all bands
    plt.ylim(core_band[0]-0.5, conduction_band[1]+0.5)
    
    # Hide x-axis
    plt.tick_params(axis='x', which='both', bottom=False, labelbottom=False)
    
    # Add legend
    plt.plot([], [], color='gold', linewidth=10, label='Occupied States')
    plt.plot([], [], color='black', alpha=0.8, linewidth=10, label='Unoccupied States')
    plt.legend(loc='upper right')
    
    plt.grid(True, axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()
```

```python
# Show band structures for different materials
print("Insulator Band Structure:")
plot_simple_band_structure(material_type="insulator")

print("Semiconductor Band Structure:")
plot_simple_band_structure(bandgap=1.5, material_type="semiconductor")

print("Conductor Band Structure:")
plot_simple_band_structure(material_type="conductor")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider

def plot_interactive_band_structure(bandgap=1.5, valence_fill=1.0):
    """
    Interactive band structure showing bands with boxes
    
    Parameters:
    bandgap - Energy difference between bands (eV)
    valence_fill - Fraction of valence band that is filled (0-1)
    """
    plt.figure(figsize=(8, 6))
    
    # Determine material type based on parameters
    if bandgap > 3.0:
        material_type = "insulator"
    elif bandgap < 0.1 or (bandgap < 0.5 and valence_fill < 0.95 and valence_fill > 0.05):
        material_type = "conductor"
    else:
        material_type = "semiconductor"
    
    # Band width will be constant
    band_width = 1.0
    
    # Box dimensions
    box_width = 0.4
    box_x = [-box_width/2, box_width/2]
    
    # Energy ranges for bands
    core_band = [-band_width*3 - 1, -band_width*2 - 1]
    valence_band = [-band_width, 0]
    conduction_band = [bandgap, bandgap + band_width]
    
    # Calculate fill heights
    valence_fill_height = valence_band[0] + (valence_band[1] - valence_band[0]) * valence_fill
    
    # Draw core band (always fully filled)
    plt.fill_between(box_x, [core_band[0], core_band[0]], [core_band[1], core_band[1]], 
                     color='gold', edgecolor='black', linewidth=1.5, zorder=3)
    
    # Draw valence band - filled portion
    plt.fill_between(box_x, [valence_band[0], valence_band[0]], [valence_fill_height, valence_fill_height], 
                     color='gold', edgecolor=None, linewidth=0, zorder=3)
    
    # Draw valence band - unfilled portion
    if valence_fill < 1.0:
        plt.fill_between(box_x, [valence_fill_height, valence_fill_height], [valence_band[1], valence_band[1]], 
                        color='black', alpha=0.8, edgecolor=None, linewidth=0, zorder=3)
    
    # Draw valence band outline
    plt.plot([box_x[0], box_x[0]], [valence_band[0], valence_band[1]], 'k-', linewidth=1.5, zorder=4)
    plt.plot([box_x[1], box_x[1]], [valence_band[0], valence_band[1]], 'k-', linewidth=1.5, zorder=4)
    plt.plot([box_x[0], box_x[1]], [valence_band[0], valence_band[0]], 'k-', linewidth=1.5, zorder=4)
    plt.plot([box_x[0], box_x[1]], [valence_band[1], valence_band[1]], 'k-', linewidth=1.5, zorder=4)
    
    # Draw conduction band 
    conduction_fill = 0
    if material_type == "conductor" and bandgap < 0.1:
        # For true conductors with overlapping bands, show partial filling
        conduction_fill = 1.0 - valence_fill
        conduction_fill_height = conduction_band[0] + (conduction_band[1] - conduction_band[0]) * conduction_fill
        
        # Draw filled portion of conduction band
        plt.fill_between(box_x, [conduction_band[0], conduction_band[0]], 
                         [conduction_fill_height, conduction_fill_height], 
                         color='gold', edgecolor=None, linewidth=0, zorder=3)
        
        # Draw unfilled portion of conduction band
        plt.fill_between(box_x, [conduction_fill_height, conduction_fill_height], 
                         [conduction_band[1], conduction_band[1]], 
                         color='black', alpha=0.8, edgecolor=None, linewidth=0, zorder=3)
    else:
        # Normal unfilled conduction band
        plt.fill_between(box_x, [conduction_band[0], conduction_band[0]], 
                         [conduction_band[1], conduction_band[1]], 
                         color='black', alpha=0.8, edgecolor=None, linewidth=0, zorder=3)
    
    # Draw conduction band outline
    plt.plot([box_x[0], box_x[0]], [conduction_band[0], conduction_band[1]], 'k-', linewidth=1.5, zorder=4)
    plt.plot([box_x[1], box_x[1]], [conduction_band[0], conduction_band[1]], 'k-', linewidth=1.5, zorder=4)
    plt.plot([box_x[0], box_x[1]], [conduction_band[0], conduction_band[0]], 'k-', linewidth=1.5, zorder=4)
    plt.plot([box_x[0], box_x[1]], [conduction_band[1], conduction_band[1]], 'k-', linewidth=1.5, zorder=4)
    
    # Add band labels
    plt.text(box_width*0.7, core_band[0] + (core_band[1] - core_band[0])/2, 
             "Core\nBand", ha='left', va='center')
    plt.text(box_width*0.7, valence_band[0] + (valence_band[1] - valence_band[0])/2, 
             "Valence\nBand", ha='left', va='center')
    plt.text(box_width*0.7, conduction_band[0] + (conduction_band[1] - conduction_band[0])/2, 
             "Conduction\nBand", ha='left', va='center')
    
    # Label the bandgap if it exists
    if bandgap > 0.1:
        plt.annotate('', xy=(0, conduction_band[0]), xytext=(0, valence_band[1]),
                    arrowprops=dict(arrowstyle='<->', color='red', lw=2))
        plt.text(box_width*0.7, valence_band[1] + bandgap/2, f"Bandgap\n{bandgap:.1f} eV", 
                 ha='left', va='center', color='red')
    
    # Calculate Fermi level
    if valence_fill < 1.0:
        fermi_level = valence_fill_height
    elif bandgap < 0.1:  # Bands overlap
        fermi_level = valence_band[1]
    else:
        fermi_level = valence_band[1] + bandgap/2
    
    # Label Fermi level
    plt.axhline(y=fermi_level, color='blue', linestyle='--', alpha=0.7)
    plt.text(-box_width, fermi_level, "Fermi\nLevel", ha='right', va='center', color='blue')
    
    # Set labels and title
    plt.title(f"{material_type.capitalize()} Band Structure\nBandgap: {bandgap:.2f} eV, Valence Fill: {valence_fill*100:.0f}%")
    plt.ylabel("Energy (eV)")
    plt.xlim(-box_width*2, box_width*2)
    
    # Adjust y limits to show all bands
    plt.ylim(core_band[0]-0.5, conduction_band[1]+0.5)
    
    # Hide x-axis
    plt.tick_params(axis='x', which='both', bottom=False, labelbottom=False)
    
    # Add legend
    plt.plot([], [], color='gold', linewidth=10, label='Occupied States')
    plt.plot([], [], color='black', alpha=0.8, linewidth=10, label='Unoccupied States')
    plt.legend(loc='upper right')
    
    plt.grid(True, axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()

def interactive_band_structure():
    """Create interactive widgets to control band structure parameters"""
    interact(
        plot_interactive_band_structure,
        bandgap=FloatSlider(min=0.0, max=5.0, step=0.1, value=1.5, 
                           description='Bandgap (eV):'),
        valence_fill=FloatSlider(min=0.0, max=1.0, step=0.05, value=1.0, 
                                description='Valence Fill:')
    )
```
