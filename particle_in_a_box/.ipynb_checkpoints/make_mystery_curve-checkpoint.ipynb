```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# True Gaussian parameters
mu = 0       # mean
sigma = 1    # standard deviation
amplitude = 1  # amplitude

# Create x values
x = np.linspace(-5, 5, 100)

# Generate the true Gaussian curve
y_true = amplitude * np.exp(-0.5 * ((x - mu) / sigma)**2)

# Add noise (random Gaussian noise)
noise = np.random.normal(0, 0.1, size=y_true.shape)
y_noisy = y_true + noise

# Create a DataFrame to store the x and noisy y values
data = pd.DataFrame({'x': x, 'y': y_noisy})

# Save the data to a CSV file
data.to_csv('mystery_curve.csv', index=False)

# Plot the noisy data and the true Gaussian curve
plt.scatter(x, y_noisy, color='red', label='Noisy Data')
plt.plot(x, y_true, label='True Gaussian Curve', color='blue', linewidth=2)
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Noisy Gaussian Curve for Fitting')
plt.show()

print("Data saved to 'noisy_gaussian_data.csv'")
```
