```python
def QHO_Hamiltonian(wavefunction):
    psi = sp.Symbol('psi')
    omega = sp.Symbol('omega')
    x = sp.Symbol('x')
    T = - sp.Rational(1,2) * sp.Derivative(psi,x,2)
    V = (sp.Rational(1,2) * omega ** 2  * x ** 2 ) * psi
    H = T + V
    H_eval = H.subs(psi,test_psi)
    eigenvalue = H_eval.doit().expand() / wavefunction
    print(eigenvalue)
    return H_eval
```

```python
test_psi = sp.exp(-omega * x**2/2)
test_psi
honpsi = QHO_Hamiltonian(test_psi)
honpsi.doit().expand()
```

```python
def plot_hamiltonian(wavefunction):
    psi = sp.Symbol('psi')
    omega = sp.Symbol('omega')
    x = sp.Symbol('x')
    T_psi = - sp.Rational(1,2) * sp.Derivative(psi,x,2)
    T_psi_eval = sp.lambdify(x,T_psi.subs(psi,wavefunction).doit().expand().subs(omega,1))
    
    V = (sp.Rational(1,2) * omega ** 2  * x ** 2 )
    V_psi = V * psi
    V_psi_eval = sp.lambdify(x,V_psi.subs(psi,wavefunction).doit().expand().subs(omega,1))
    
    H_psi = T_psi + V_psi
    H_psi_eval = sp.lambdify(x,H_psi.subs(psi,wavefunction).doit().expand().subs(omega,1))
   
    x_array = np.linspace(-5,5,1000)
    
    plt.plot([], [], ' ', label="Extra label on the legend")
    
    y_line = np.zeros_like(x_array)
    plt.plot(x_array,y_line, color=r'grey')
    
    potential_array = V_psi_eval(x_array)
    plt.plot(x_array,potential_array,color = r'grey',label=r'$V(r)$')

    wavefunction_eval = sp.lambdify(x,wavefunction.subs(omega,1))
    y_wfn = wavefunction_eval(x_array)
    plt.plot(x_array,y_wfn,color='black',label=r'$\psi(r)$')

    y_operated_wfn = H_psi_eval(x_array)
    plt.plot(x_array,y_operated_wfn,color='red',label=r'$\hat{H}\psi(r)$')

    y_pot_wfn = V_psi_eval(x_array)
    plt.plot(x_array,y_pot_wfn,color='orange',alpha=0.5,label=r'$V(r)\psi(r)$')

    y_ke_wfn = T_psi_eval(x_array)
    plt.plot(x_array,y_ke_wfn,color='blue',alpha=0.3,label=r'$\hat{T}\psi(r)$')

    plt.legend()
    
    plt.show()
```

```python
plot_hamiltonian(test_psi)
```

```python
def check_eigenvalue(psi_trans,psi):
    _lambda = sp.Symbol('lambda')
    equation = sp.Eq(psi_trans, _lambda * psi)
    solution = sp.solve(equation)
    return solution
```

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as axes3d
import matplotlib.colors as mcolors
from scipy.special import sph_harm
import matplotlib.cm as cm

def plot_spherical_harmonic(l, m):
    '''
    Adapted from code at https://github.com/DalInar/schrodingers-snake
    '''
    thetas = np.linspace(0, np.pi, 50)
    phis = np.linspace(0, 2 * np.pi, 50)
    
    Theta, Phi = np.meshgrid(thetas, phis)
    s_harm = sph_harm(m, l, Phi, Theta)  # Spherical harmonic function

    R = np.abs(s_harm)  # Magnitude
    X = R * np.sin(Theta) * np.cos(Phi)
    Y = R * np.sin(Theta) * np.sin(Phi)
    Z = R * np.cos(Theta)

    # Compute phase and normalize to [0,1]
    phase = np.angle(s_harm)
    phase_norm = (phase + np.pi) / (2 * np.pi)  # Normalize from [-, ] to [0,1]

    # Create colormap based on phase
    cmap = cm.hsv  # HSV is good for phase representation
    colors = cmap(phase_norm)  # Map phase to colors
    
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(1, 1, 1, projection='3d')
    surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, facecolors=colors, linewidth=0, antialiased=False, alpha=0.8)

    # Create colorbar manually
    sm = cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=-np.pi, vmax=np.pi))
    sm.set_array([])  # Dummy array for colorbar
    cbar = plt.colorbar(sm, shrink=0.6, pad=0.1, ax=ax)

    # Set custom phase labels
    cbar.set_label("Phase")
    cbar.set_ticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
    cbar.set_ticklabels([r'$-1$', r'$-i$', r'$1$', r'$i$', r'$-1$'])

    # Labels and title
    plt.title(r'Solutions to Rigid Rotor Wavefunction, $Y_l^m(\theta,\phi)$' + r', $l=$' + str(l) + r', $m=$' + str(m))
    ax.set_xlabel(r'$x$')
    ax.set_ylabel(r'$y$')
    ax.set_zlabel(r'$z$')
    
    plt.show()
```

```python
# Example usage:
plot_spherical_harmonic(10, 2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as axes3d
import matplotlib.colors as mcolors
from scipy.special import sph_harm
import matplotlib.cm as cm

def plot_rigid_rotor(J, m):
    '''
    Adapted from code at https://github.com/DalInar/schrodingers-snake
    '''
    thetas = np.linspace(0, np.pi, 50)
    phis = np.linspace(0, 2 * np.pi, 50)
    
    Theta, Phi = np.meshgrid(thetas, phis)
    s_harm = sph_harm(m, J, Phi, Theta) 

    R = np.abs(s_harm)  
    X = R * np.sin(Theta) * np.cos(Phi)
    Y = R * np.sin(Theta) * np.sin(Phi)
    Z = R * np.cos(Theta)

    phase = np.angle(s_harm)
    phase_norm = (phase + np.pi) / (2 * np.pi) 

    cmap = cm.hsv
    colors = cmap(phase_norm) 
    
    fig = plt.figure(figsize=(9, 8)) 
    ax = fig.add_subplot(1, 1, 1, projection='3d')
    surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, facecolors=colors, linewidth=0, antialiased=False, alpha=0.8)

    sm = cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=-np.pi, vmax=np.pi))
    sm.set_array([])
    cbar = plt.colorbar(sm, shrink=0.6, pad=0.05, ax=ax)

    cbar.set_label("Phase")
    cbar.set_ticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
    cbar.set_ticklabels([r'$-1$', r'$-i$', r'$1$', r'$i$', r'$-1$'])

    energy = J * (J + 1) 
    title_text = (rf'Rigid Rotor Wavefunction ( J={J}, m={m} )')
    plt.figtext(0.75, 0.20, f'$E_J = J (J+1) \\frac{{\\hbar^2}}{{2I}}= {energy} \\frac{{\\hbar^2}}{{2I}}$', ha='center', fontsize=12,color='red',
            bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

    plt.suptitle(title_text, fontsize=18, ha='center',y=0.85)

    fig = plt.gcf()
    fig.canvas.draw()  
    line = plt.Line2D([0.18, 0.82], [0.81 , 0.81 ], 
                      color='black', linewidth=1, transform=fig.transFigure, clip_on=False)
    fig.add_artist(line)
    
    ax.set_xlabel(r'$x$')
    ax.set_ylabel(r'$y$')
    ax.set_zlabel(r'$z$')

    plt.show()
```

```python

# Example usage:
plot_rigid_rotor(2, 2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact, FloatSlider

def plot_spherical_vector(r, theta, phi):
    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)
    
    fig = plt.figure(figsize=(4,4))
    ax = fig.add_subplot(111, projection='3d')
    
    u = np.linspace(0, 2*np.pi, 20)
    v = np.linspace(0, np.pi, 20)
    xs = r * np.outer(np.cos(u), np.sin(v))
    ys = r * np.outer(np.sin(u), np.sin(v))
    zs = r * np.outer(np.ones_like(u), np.cos(v))
    ax.plot_surface(xs, ys, zs, rstride=2, cstride=2, color='cyan', alpha=0.1, 
                    edgecolor='black', linewidth=0.5)

    ax.plot([0, x], [0, y], [0, z], color='red', linewidth=2, label='r')
    ax.scatter(x, y, z, color='red', s=10)
    
    r_phi = 0.5  
    phi_arc = np.linspace(0, phi, 100)
    x_phi_arc = r_phi * np.cos(phi_arc)
    y_phi_arc = r_phi * np.sin(phi_arc)
    z_phi_arc = np.zeros_like(phi_arc)
    ax.plot(x_phi_arc, y_phi_arc, z_phi_arc, color='blue', label=r'$\phi$')
    
    r_theta = 0.5  
    theta_arc = np.linspace(0, theta, 100)
    x_theta_arc = r_theta * np.sin(theta_arc) * np.cos(phi)
    y_theta_arc = r_theta * np.sin(theta_arc) * np.sin(phi)
    z_theta_arc = r_theta * np.cos(theta_arc)
    ax.plot(x_theta_arc, y_theta_arc, z_theta_arc, color='green', label=r'$\theta$')

    max_range = 1.5
    ax.plot([0, max_range], [0, 0], [0, 0], color='black', linestyle=':')
    ax.plot([0, 0], [0, max_range], [0, 0], color='black', linestyle=':')
    ax.plot([0, 0], [0, 0], [0, max_range], color='black', linestyle=':')
    ax.set_xlim([-max_range, max_range])
    ax.set_ylim([-max_range, max_range])
    ax.set_zlim([-max_range, max_range])

    ax.legend()
    ax.set_title("Spherical Coordinates")
    
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    
    plt.show()
```

```python
interact(plot_spherical_vector, 
         r=FloatSlider(min=0.1, max=1.5, step=0.1, value=1, description='r'),
         theta=FloatSlider(min=0, max=np.pi, step=0.1, value=np.pi/4, description=r'$\theta$'),
         phi=FloatSlider(min=0, max=2*np.pi, step=0.1, value=np.pi/4, description=r'$\phi$'))
```
