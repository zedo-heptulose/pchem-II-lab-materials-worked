```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def delta_E(gamma,alpha,B_0):
    return gamma * (1 - alpha) * B_0
```

```python
delta_E(20,0.5,0.5)
```

```python
def plot_stuff(B_0):
    x_vals = np.linspace(0,10,100)
    y_vals = np.linspace(0,1,100)
    X,Y = np.meshgrid(x_vals, y_vals)
    Bnot = B_0
    Z = delta_E(X,Y,Bnot)
    fig = plt.figure(figsize=(4,4))
    ax = fig.add_subplot(111,projection='3d')
    
    ax.plot_surface(X,Y,Z, cmap='viridis')
    
    ax.set_xlabel('$\\gamma$')
    ax.set_ylabel('$\\alpha$')
    # ax.set_zlabel('f(x,y)')
    ax.set_title('$\\Delta{}E$')
    # plt.gca.invert_axis()
    ax.set_zlim(0,200)
    
    plt.tight_layout()
    
    plt.show()
```

```python
#I mean pack it up right? we did it lol
```

```python
plot_stuff(0)
```
